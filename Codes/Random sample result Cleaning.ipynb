This code is to cleaning the raw data from random sampled experiment

```python
#import experiment data
###very useful
import pandas as pd
import glob

# path: where you store the raw data, it should be csv file
path = r'' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
data=frame[['participant','img1','img2','key_resp_4.keys']]
```

```python
result={}
for i in data.iloc():
    if i['participant'] not in result.keys():
        result[i['participant']]={}
    if i['key_resp_4.keys']=='q':
        result[i['participant']][f'{i["img1"]}']=1
        result[i['participant']][f'{i["img2"]}']=0
    else:
        result[i['participant']][f'{i["img1"]}']=0
        result[i['participant']][f'{i["img2"]}']=1
```
